```python
# Wanted Characteristics:

# Carat: 0.25
# cut: Premium
# color: J
# depth: 56.7
# table: 57.0
# x: 4.01
# y: 4.05
# z: 2.35
```

```python
# importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns

#dataset names from seaborn library
sns.get_dataset_names()
```

```python
#importing diamonds dataset
diamonds_df = sns.load_dataset("diamonds")

#print dataset header
diamonds_df.head()

```

```python
#print dataset tail
diamonds_df.tail()
```

## Divide Data into Features and Labels

```python
#extracting features 
X = diamonds_df.drop(['price'], axis=1)

#extracting labels 
y = diamonds_df["price"]

#printing features
X.head()
```

```python
#print the labels
y.head()
```

## Convert Categorical Data to Numbers

```python
# removing categorical features
numerical = X.drop(['cut','color','clarity'], axis = 1)

#print numeric features only
numerical.head()
```

```python
# filter categorical features
categorical = X.filter(['cut','color',])

#print categorical features header
categorical.head()
```

```python
#one-hot encoding
cat_numerical = pd.get_dummies(categorical, drop_first = True)
cat_numerical.head()
```

```python
#concatenate numeric and one hot-encoded features

X = pd.concat([numerical, cat_numerical], axis = 1)
X.head()
```

```python
## divide data into Training and Test Datasets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

## Data Scaling/Normalization

```python
# feature scaling using standard scaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Linear Regression

```python
#import linear regression model 
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

#training the model
regressor = lin_reg.fit(X_train, y_train)

#making predictions on the test set
y_pred = regressor.predict(X_test)

#evaluating model performance
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
